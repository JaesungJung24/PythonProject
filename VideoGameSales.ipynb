```python
from collections import Counter

fig = pl.figure()
ax = fig.add_axes([0,0,1,1])
genre = [x for x in Counter(data['Genre']).keys()]
count = [x for x in Counter(data['Genre']).values()]
pl.xticks(rotation=50)



ax.bar(genre, count)
pl.savefig('bar_graph.png',bbox_inches='tight', dpi=150)
pl.show()
```

```python
from statistics import mean
fig = pl.figure(figsize =(10, 7))
ax = fig.add_axes([0,0,1,1])
global_sales = []
year = sorted([int(x) for x in Counter(data['Year']).keys()])
for i in year:
  global_sales.append(data['Global_Sales'].loc[data.Year == i])
pl.xticks(rotation=50)
ax.boxplot(global_sales)
xtickNames = pl.setp(ax, xticklabels=year)
pl.ylim((0, 40))
pl.savefig('box_plot.png',bbox_inches='tight', dpi=150)
pl.show()

# print(year)
# print(global_sales)
# print(len(year), len(global_sales))
# print(data.shape)
```

```python
from statistics import mean
fig = pl.figure()
ax = fig.add_axes([0,0,1,1])
global_sales = []
year = sorted([int(x) for x in Counter(data['Year']).keys()])
for i in year:
  global_sales.append(data['Global_Sales'].loc[data.Year == i])
global_sales_mean = [mean(x) for x in global_sales]
pl.xticks(rotation=50)

ax.bar(year, global_sales_mean)
pl.savefig('mean_graph.jpg',bbox_inches='tight', dpi=150)
pl.show()


#print(year)
#print(global_sales)
#print(len(year), len(global_sales))
#print(data.shape)
```

```python
# Drawing multiple plots in one plot using same x,y axis

fig = pl.figure()
ax = fig.add_axes([0,0,1,1])
global_sales = []
na_sales = []
rank_lst = []
year = sorted([int(x) for x in Counter(data['Year']).keys()])
for i in year:
  global_sales.append(data['Global_Sales'].loc[data.Year == i])
  na_sales.append(data['NA_Sales'].loc[data.Year == i])
  rank_lst.append(data['Rank'].loc[data.Year == i])

global_sales_mean = [mean(x) for x in global_sales]
na_sales_mean = [mean(x) for x in na_sales]
rank_lst_mean = [mean(x) for x in rank_lst]

pl.xticks(rotation=50)
# ax.bar(year, global_sales_mean)
pl.plot(year, global_sales_mean, label='Global Sales')
pl.plot(year, na_sales_mean, label="NA Sales")
pl.legend(loc="upper right")
ax.set_xlabel("Year", fontsize = 14)
ax.set_ylabel("Sales", fontsize = 14)

pl.savefig('filename.jpg',bbox_inches='tight', dpi=150)
pl.show()
```

```python
# Drawing multiple plots in one plot using two y axes

# create figure and axis objects with subplots()
fig,ax = pl.subplots()
ax.plot(year,
        global_sales_mean,
        color="red", 
        marker="o")
ax.set_xlabel("Year", fontsize = 14)
ax.set_ylabel("Global Sales",
              color="red",
              fontsize=14)
# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(year, rank_lst_mean ,color="blue",marker="o")
ax2.set_ylabel("Rank",color="blue",fontsize=14)
pl.show()
# save the plot as a file
# fig.savefig('two_different_y_axis_for_single_python_plot_with_twinx.jpg',
#             format='jpeg',
#             dpi=100,
#             bbox_inches='tight')
```
